```
# import the necessary packages
import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt

# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

```
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

**Horizontal Edge Detector** : Edge detection is simply achieved by finding the regions in a image which has sharp intensity (high brightness value) or sharp change in color in the image. Basically, it is a gradient based approach. In horizontal edge detector, we highlight the horizontal streaks that have high brightness value. 

```
# Horizontal edge detector
kernel = np.float32([[-1,-1,-1],[0,0,0],[1,1,1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

** 45 Degree Angle Edge Detector** : To extract a angled edge, we basically extract a diagonal kernel using diagonal matrix and then we multiply the original matrix with the diagonal kernel we obatined earlier.

```
# 45 degree angle edge detector
kernel = np.float32([[1,-1,-1],[0,1,-1],[0,0,1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

**Blur Kernel** : Blur is often obtained by removing noises from an image. Blur kernel is obtained by convolving the original image with the kernel that is obtained after removing the noise from the image.

```
# Blur kernal
kernel = np.float32([[0.140,0.120,0.140],[0.140,0.120,0.140],[0.140,0.120,0.140]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
#cv2_imshow(image)
```

**Sharpen Kernel** : Sharpening of a image is obtained by enhancing the edges of a image with the help of a special defined kernel matrix of the original image.

```
# Sharpen kernal
kernel = np.float32([[0,-1,0],[-1,5,-1],[0,-1,0]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
#cv2_imshow(image)
```

**Identity Kernel** : Identity image is obtained by convolving the original image with a kernel which nullifies any change that is brought upon the neighboring
 pixels present in the image.


```
# Identity Kernel
kernel = np.float32([[ 0,  0,  0],[0,  1,  0],[0,  0,  0]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```
